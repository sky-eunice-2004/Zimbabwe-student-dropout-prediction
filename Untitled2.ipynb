#  Data Preprocessing

```python
# Select only important features (for Flask compatibility)
X = df[['age', 'attendance_rate', 'study_hours', 'stress_level']]
y = df['dropout']
```

```python
X = df[['age', 'attendance_rate', 'study_hours', 'stress_level']].copy()
```

```python
#clean text

X.loc[:, 'stress_level'] = X['stress_level'].astype(str)
X.loc[:, 'stress_level'] = X['stress_level'].str.strip().str.lower()
```

```python
#covert to numbers
X['stress_level'] = X['stress_level'].map({
    'Low': 1,
    'Moderate': 2,
    'High': 3
})
```

```python
X['stress_level'] = X['stress_level'].fillna(2)
```

```python
#handling missing values 

X = X.replace([np.inf, -np.inf], np.nan)
X = X.fillna(X.mean())
```
